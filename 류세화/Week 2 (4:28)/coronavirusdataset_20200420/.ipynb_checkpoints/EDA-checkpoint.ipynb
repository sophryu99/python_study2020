## 날짜변수 Split

```python
# 년, 월, 일 Split 함수 만들기
def parse_month(date):
    month = date.split('-')[1]
    month = int(month)
    return month

parse_month('2020-01-21')
```

```python
def parse_date(date):
    day = date.split('-')[-1]
    day = int(day)
    return day

parse_date('2020-01-21')
```

```python
# 월, 일 칼럼 붙이기
dropped['month']=dropped['date'].apply(parse_month)
dropped['day'] = dropped['date'].apply(parse_date)
dropped.drop(columns = 'date')
```
