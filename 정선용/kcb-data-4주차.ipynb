```python
import pandas as pd
import seaborn as sns
import warnings
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rc("font",family="Malgun Gothic")
```

```python
df2=pd.read_csv("data/2_지역단위직업군.CSV")
df2
```

```python
#나이대에 따라 사는 거주지 비교

g=pd.pivot_table(df2,index=["CT_PRVNC_NM","CT_CNTY_GU_NM"],values="AGE_CD").sort_values(by="AGE_CD").plot.barh(figsize=(25,10)).set_title("나이에 따른 거주지")

```

```python
#대기업,일반기업,전문직,대표 소득자 비율 비교
plt.figure(figsize=(15,10))
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_410",color="black",ci=0).set_title("대기업")
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_420",color="red",ci=0).set_title("일반기업")
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_430",color="green",ci=0).set_title("전문직")
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_440",ci=0).set_title("대기업,일반기업,전문직,대표 소득자 비율")
```

```python
#일반자영업자,전문직자영업자,기타 소득비율 비교
plt.figure(figsize=(15,10))
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_510",color="black",ci=0).set_title("일반자영업자")
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_520",color="red",ci=0).set_title("전문직자영업자")
sns.lineplot(data=df2,x="CT_PRVNC_NM",y="RT_MDL_GB_910",color="green",ci=0).set_title("일반자영업자,전문직자영업자,기타 소득자 비율")
```

```python
#지역단위 소득 
df3=pd.read_csv("data/3_지역단위소득.CSV")
df3
```

```python
#성별에 따른 평균연소득과 상위연소득 비교
a=pd.pivot_table(data=df3,index="성별",values=["평균연소득","상위연소득"])
a.plot.bar().set_title("성별에 따른 평균연소득과 상위연소득 비교")
```
